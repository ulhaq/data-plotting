# Task 1

You might need to enlarge the chart to see the data

```python
%matplotlib notebook
import kkdata
import pprint
import numpy as np
import matplotlib.pyplot as plt

rs = {}

for year in kkdata.STATISTICS:
    for city in kkdata.STATISTICS[year]:
        for age in kkdata.STATISTICS[year][city]:
            for code in kkdata.STATISTICS[year][city][age]:
                if code == '5170' or code == '5180':
                    if not age in rs:
                        rs[age] = {}

                    if not 'britain' in rs[age]:
                        rs[age]['britain'] = 0
                    if code == '5170':
                        rs[age]['britain'] += int(kkdata.STATISTICS[year][city][age][code])

                    if not 'germany' in rs[age]:
                        rs[age]['germany'] = 0
                    if code == '5180':
                        rs[age]['germany'] += int(kkdata.STATISTICS[year][city][age][code])
brt = []
ger = []
for i in rs.values():
    brt.append(i['britain'])
    ger.append(i['germany'])

n_groups = len(rs)
german = ger
british = brt

plt.subplots()
index = np.arange(n_groups)
bar_width = 0.35
opacity = 0.8
 
plt.bar(index, german, bar_width,alpha=opacity,color='b',label='German')
 
plt.bar(index + bar_width, british, bar_width,alpha=opacity,color='g',label='British')
 
plt.title('Citizens in terms of ages')
plt.xlabel('Ages')
plt.ylabel('Citizens')
plt.xticks(index + bar_width, rs.keys())
plt.legend()
 
plt.tight_layout()
plt.show()
```

# Task 3

This will plot a pie chart for the five major citizen groups (excluding danish citizens) of age between 20-65

```python
import kkdata
import pprint
import numpy as np
import matplotlib.pyplot as plt

rs = {}

for year in kkdata.STATISTICS:
# year = '2015'
    for city in kkdata.STATISTICS[year]:
        for age in kkdata.STATISTICS[year][city]:
            for code in kkdata.STATISTICS[year][city][age]:
                if not code in rs:
                    rs[code] = 0
                rs[code] += int(kkdata.STATISTICS[year][city][age][code])

srt_val = sorted(rs.items(), key=lambda kv: kv[1], reverse=True)

labels = []
sizes = []

for code, citizens in srt_val[1:6]:
    labels.append(code)
    sizes.append(citizens)

explode = (0.01, 0.01, 0.01, 0.01, 0.01)

fig1, ax = plt.subplots()
ax.set_title('Five major citizen groups of age between 20-65')
ax.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%', startangle=90)

plt.show()
```
